# Exploratory Data Analysis - Credit Card Fraud Detection
## Credit Card Approval Prediction System

This notebook performs comprehensive EDA on the Credit Card Fraud Detection dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (12, 8)
print('Libraries imported successfully')
```

```python
df = pd.read_csv('../dataset/creditcard.csv')
print(f'Dataset shape: {df.shape}')
df.head()
```

```python
print('Dataset Info:')
df.info()
```

```python
print('\nMissing Values:', df.isnull().sum().sum())
print('Duplicate Rows:', df.duplicated().sum())
print('\nClass Distribution:')
print(df['Class'].value_counts())
print(f'\nFraud Percentage: {df["Class"].mean()*100:.4f}%')
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('EDA - Credit Card Fraud Detection', fontsize=18, fontweight='bold')

sns.countplot(x='Class', data=df, ax=axes[0,0], palette=['#2ecc71', '#e74c3c'])
axes[0,0].set_title('Class Distribution')
axes[0,0].set_xticklabels(['Legitimate (0)', 'Fraudulent (1)'])

axes[0,1].hist([df[df['Class']==0]['Amount'], df[df['Class']==1]['Amount']],
               bins=50, label=['Legitimate', 'Fraudulent'], alpha=0.7,
               color=['#2ecc71', '#e74c3c'])
axes[0,1].set_title('Transaction Amount Distribution')
axes[0,1].set_xlabel('Amount')
axes[0,1].set_ylabel('Frequency')
axes[0,1].legend()

corr = df.drop('Class', axis=1).corr()
sns.heatmap(corr, ax=axes[0,2], cmap='RdBu', center=0, cbar_kws={'shrink': 0.8})
axes[0,2].set_title('Correlation Heatmap')

df.boxplot(column='Amount', by='Class', ax=axes[1,0])
axes[1,0].set_title('Amount by Class')
axes[1,0].set_xlabel('Class')

for col in ['V1', 'V2', 'V3', 'V4', 'V5']:
    sns.kdeplot(df[df['Class']==0][col], ax=axes[1,1], color='#2ecc71', alpha=0.5)
    sns.kdeplot(df[df['Class']==1][col], ax=axes[1,1], color='#e74c3c', alpha=0.5)
axes[1,1].set_title('Feature Distributions (V1-V5)')
axes[1,1].set_xlabel('Value')

class_counts = df['Class'].value_counts()
axes[1,2].pie(class_counts, labels=['Legitimate', 'Fraudulent'],
              autopct='%1.2f%%', startangle=90, colors=['#2ecc71', '#e74c3c'],
              explode=(0, 0.05))
axes[1,2].set_title('Class Distribution Pie')

plt.tight_layout()
plt.savefig('../reports/eda_visualizations.png', dpi=150, bbox_inches='tight')
plt.show()
```
